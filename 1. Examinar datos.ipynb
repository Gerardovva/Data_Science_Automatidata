```python
# ¿Cuántos de cada tipo de pago están representados en los datos?
df["payment_type"].value_counts()
```

According to the data dictionary, the payment method was encoded as follows:

1 = Credit card  
2 = Cash  
3 = No charge  
4 = Dispute  
5 = Unknown  
6 = Voided trip

Según el diccionario de datos, el método de pago se codificó de la siguiente manera:

1 = Tarjeta de crédito  
2 = Efectivo  
3 = Sin cargo  
4 = Impugnación  
5 = Desconocido  
6 = Viaje anulado

```python
# What is the average tip for trips paid for with credit card?
# ¿Cuál es la propina media de los viajes pagados con tarjeta de crédito?
viajes_tarjeta=df[df["payment_type"]==1]
propina_media_tarjeta = viajes_tarjeta['tip_amount'].mean()
propina_media_tarjeta=round(propina_media_tarjeta,2)
print(f"La propina media de tarjeta es: { propina_media_tarjeta} dolares")

# What is the average tip for trips paid for with cash?
# ¿Cuál es la propina media de los viajes pagados en efectivo?

propina_efectivo = df[df['payment_type']==2]['tip_amount'].mean()
print(f'La propina media de efectivo es: {propina_efectivo} dolares')
```

```python
# How many times is each vendor ID represented in the data?
#¿Cuántas veces está representado en los datos el ID de cada proveedor?
df['VendorID'].value_counts()
```

```python
# What is the mean total amount for each vendor?
# ¿Cuál es el monto total medio de cada proveedor?

df.groupby("VendorID").mean()[["total_amount"]]
```

```python

# Filtrar los datos solo para pagos con tarjeta de crédito
credit_card= df[df['payment_type'] ==1]

# Filtrar los datos de la tarjeta de crédito únicamente para el recuento de pasajeros
credit_card['passenger_count'].value_counts()
```

```python
# Calcule el importe medio de la propina para cada pasajero (sólo pagos con tarjeta de crédito)

credit_card.groupby("passenger_count").mean(numeric_only=True)[["tip_amount"]]

```
